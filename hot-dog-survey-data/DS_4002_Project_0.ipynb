```python
import pandas as pd

df = pd.read_excel("Hotdog Survey (Responses).xlsx")

print(df.shape)
df.head()
```

```python
for col in df.columns:
    print(repr(col))
```

```python
df = df.rename(columns={
    "Timestamp": "timestamp",
    "Year at UVA": "year",
    "What is your Gender Identity?": "gender",
    "What Schools are you a Part of?": "schools",
    "Is a Hotdog a Sandwich?": "is_sandwich",
    "How strongly do you feel about your previous answer?": "strength",
    "How frequently do you eat hot dogs?": "frequency"
})

print(df.columns.tolist())
```

```python
df["year"].value_counts(dropna=False)
```

```python
undergrad_years = ["1st Year", "2nd Year", "3rd Year", "4th Year"]

df = df[df["year"].isin(undergrad_years)]

print("Rows after undergrad filter:", len(df))
```

```python
df["is_sandwich"] = (
    df["is_sandwich"]
    .astype(str)
    .str.strip()
    .str.lower()
)

df["is_sandwich"].value_counts()
```

```python
x = (df["is_sandwich"] == "yes").sum()
n = len(df)

print("x (yes responses):", x)
print("n (total rows):", n)
print("Sample proportion:", x / n)
```

```python
from statsmodels.stats.proportion import proportions_ztest

z_stat, p_value = proportions_ztest(x, n, value=0.25)

z_stat, p_value
```
